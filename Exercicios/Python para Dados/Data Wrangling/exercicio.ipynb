```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("collected research people.csv")
```

```python
df.head()
```

```python
#remoção de coluna desnecessária
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.columns
```

```python
#criação de dicionário para renomear as colunas
novas_colunas = {'Endereerço':'Endereço',
                 'phone':'Telefone',
                 'anual_income':'Renda Anual',
                 'AAge':'Idade',
                 'n_filhos':'Filhos',
                 'Food_restrictions':'Restrição Alimentar',
                 'streaming':'Streaming', 
                 'diariamente_sun blocker':'Usa protetor solar', }

df.rename(columns=novas_colunas,inplace=True)
```

```python
#estudo da coluna de 'Nome'
df['Nome']
```

O parâmetro regex=True no método replace indica que os valores a serem substituídos devem ser tratados como expressões regulares (regex) em vez de literais de texto. Isso permite realizar substituições mais complexas e poderosas.

```python
#str = manipulação de string linha por linha
#O **\W** equivale a [^a-zA-Zà-úÀ-Ú0-9_].
df['Nome'] = df['Nome'].str.replace('[^a-zA-Zà-úÀ-Ú0-9 ]','',regex=True)
df['Nome']
```

```python
#estudo da coluna de telefone, substiuição dos caracteres especiais
df['Telefone'] = df['Telefone'].str.replace('[^0-9()+]','',regex=True)
```

```python
#loalizar os que foram substituido por asteriscos (não validos), e os vazios
df.loc[df['Telefone'].str.contains('[*]',na=False), 'Telefone'] = np.nan #atribuito do numpy para localizar not a number
```

```python
df['Telefone']
```

```python
#dicionário para renomear valores em coluna específica
estado_civil = {'S':'Solteiro',
                'V':'Viúvo',
                'C':'Casado',
                'D':'Divorciado'}

df['Estado Civil'] = df['Estado Civil'].replace(estado_civil)
df['Estado Civil']
```

```python
#eliminação das colunas que não iremos utilizar na abordagem
drop_carri = ['Restrição Alimentar',
              'Streaming',
              'Usa protetor solar',
              'Quantidade de Livros Comprados',
              'Autor Favorito']

df_carri = df.drop(columns=drop_carri)
df_carri.head(3)
```

```python
i = ['Quitada','Financiamento']
c = ['De favor', 'Aluguel']
investimento = np.where(df_carri['Moradia'].isin(i),'Não quer investir','Quer investir')
compra = np.where(df_carri['Moradia'].isin(c),'Quer comprar','Não quer comprar')

df_carri['Investimento de imóvel'] = investimento
df_carri['Compra de imóvel'] = compra
df_carri.head()
```
