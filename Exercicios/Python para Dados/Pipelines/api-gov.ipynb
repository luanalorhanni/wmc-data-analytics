```python
headers = {"chave-api-dados":"a66539aeba5579393d90a5fa6763a215"}
```

```python
#Descobrir quantas emendas tem no ano de 2022. Serão 408 por página
15 * 408
```

DF somente da página 2. É necessário fazer para cada página.

```python
url = 'https://api.portaldatransparencia.gov.br/api-de-dados/emendas?ano=2022&pagina='
lista_urls = []
for i in range(1, 409): #pra cada item na lista do índice 2 ao 408
    lista_urls.append(url + str(i)) #adicionar a lista a url modificada com as numerações
    
lista_urls 
```

```python
#Função para aplicar a requisão em cada url
def request(url):
    response = requests.get(url, headers=headers)
    return response.json()
```

```python
#Transformar a lista em array
array_urls = np.asarray(lista_urls)
array_urls
```

```python
#Vetorizando a função
vet_req = np.vectorize(request)
array_requests = vet_req(array_urls)
```

```python
#array_requests = array_requests.tolist()
len(array_requests)
```

```python
array_requests[1]
```

```python
#Transformar os requests em uma lista unificada
df = np.concatenate(array_requests).tolist()
```

```python
df = pd.DataFrame(df)
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
#Troca de tipo de de colunas
coluna_valores = ['valorEmpenhado','valorLiquidado','valorPago','valorRestoInscrito',
       'valorRestoCancelado', 'valorRestoPago']

for coluna in coluna_valores:
    df[coluna] = df[coluna].str.replace('.','')
    df[coluna] = df[coluna].str.replace(',','.')
    df[coluna] = df[coluna].astype(np.float32)
    
df.info()
```

```python
#Separação do dataframe
df_area = df[['funcao','valorEmpenhado']]

#Agrupamento das colunas
area_empenhado = df_area.groupby('funcao').sum('valorEmpenhado')

#Para achar a porcentagem de cada função no valor empenhado, deve-se dividir pela soma total
porcentagem = (area_empenhado/df_area['valorEmpenhado'].sum())

#Transformar num data frame
porcentagem = porcentagem.reset_index()
porcentagem
```

```python
porcentagem.to_csv('gov-area-vlrempenhado-2022')
```
