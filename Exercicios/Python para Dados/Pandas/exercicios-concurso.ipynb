```python
concurso.isnull().sum() #Contagem dos valores vazios
```

```python
import matplotlib.pyplot as plt

faltosos_percentual = concurso.isnull().mean() * 100

plt.figure(figsize=(10, 6))
faltosos_percentual.plot(kind='bar', color='skyblue')
plt.title('Porcentagem de Dados Faltosos por Coluna')
plt.xlabel('Colunas')
plt.ylabel('Porcentagem de Dados Faltosos (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
